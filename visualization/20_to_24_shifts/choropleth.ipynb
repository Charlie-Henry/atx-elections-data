```python
import geopandas as gpd
import matplotlib.pyplot as plt

dest_data = gpd.read_file("../../prior_results_estimation/travis_county_20_24.geojson")

ax = dest_data.plot(column="change_dem_pct", cmap="bwr_r", legend=True, figsize=(12, 8), legend_kwds={"fmt":".0%"}, vmin=-0.15, vmax=0.15 )
ax.set_axis_off()
ax.set_title("Change in DEM Vote Share 2020 to 2024 - Travis County TX")
plt.savefig("shift_choropleth.png", dpi=800)
```
